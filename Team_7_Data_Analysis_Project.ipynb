* *Revenu :* Existe-t-il des niveaux de revenus spécifiques qui sont associés à la possession d'un vélo?

*  *Distance de trajet :* Existe-t-il une corrélation entre la distance du trajet domicile-travail et la possession d'un vélo?

* *Gender-Âge :* Les jeunes hommes achètent-ils plus que les jeunes femmes?

* *Gender-Revenu :* Les revenus élevés des hommes ont-ils un impact plus fort sur les achats que ceux des femmes?

* *Statut-Revenu :* Comment le revenu et le statut familial affectent-ils la propension à acheter?

* *Région-Âge :* Y a-t-il des différences dans les préférences d'âge entre les régions ?

* *Région-Revenu :* Des revenus élevés dans une région donnée peuvent-ils conduire à un plus grand nombre d'achats?

# **2. Nettoyage des données**

Pour garantir une analyse précise et fiable, il est essentiel de disposer de données propres. Cette étape implique plusieurs actions clés : la vérification des doublons; la gestion des valeurs manquantes; l'élimination des données redondantes; la normalisation des données et la détection des valeurs aberrantes.

```python
# Correction d'une faute de frappe
data = data.rename(columns={'Martial Status': 'Marital Status'})
print(data.head())
```

```python
# Supprimer la colonne contenant des données inutiles
data = data.drop(columns=['ID'])
print(data.columns)
```

```python
# Déterminer le nombre de doublons du DataFrame et les supprimer
print(data.duplicated().sum(), 'doublons')
print(data.shape)

data = data.drop_duplicates()
print(data.shape)
```

```python
# Identifier les données manquantes dans le DataFrame et travailler dessus
print(data.isnull().sum())
```

```python
# Vérifier la cohérence des noms de colonnes et des types de données
data.dtypes
```

```python
# Convertissons la colonne «Income» du type objet au type interger
# data['Income'] = data['Income'].str.replace('$', '').str.replace(',','')
print(data.head())
data['Income'] = data['Income'].astype(int)
print(data.dtypes)
```

```python
# Détection des valeurs aberrantes

# Calculer des Q1, Q3 et IQR
Q1 = data['Age'].quantile(0.25)  # Замените 'Age' на интересующий вас столбец
Q3 = data['Age'].quantile(0.75)
IQR = Q3 - Q1

# Définir les limites des émissions
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

outliers = data[(data['Age'] < lower_bound) | (data['Age'] > upper_bound)]
print(outliers)

sns.boxplot(data = data, x = "Age")
plt.show()
```

# **3. Analyse de données exploratoires**

Refléter les tendances des données dans les caractéristiques catégorielles sous forme de visualisation. Ces visualisations permettront de mieux comprendre la distribution des données dans chaque colonne **de données catégorielles**.

```python
# Sélectionner des caractéristiques catégorielles et les stocker dans une nouvelle variable
cat_obj = data.select_dtypes(include=['object'])
print(cat_obj.columns)
```

```python
# Créer des visualisations
for col in cat_obj:
    freq = data[col].value_counts().reset_index()
    freq.columns = [col, 'Frequency']

    plt.figure(figsize = (8, 6))
    sns.set_style('whitegrid', {'grid.color': 'lightgray', 'grid.linestyle': '-', 'grid.alpha': 0.3})
    sns.barplot(data = freq, x = col, y = 'Frequency', hue = col, palette = 'pastel', legend = False)

    # Afficher les valeurs au-dessus des colonnes
    for index, value in enumerate(freq['Frequency']):
        plt.text(index, value, f'{value}', ha = 'center', va = 'bottom')

    plt.title(f'Frequency on column {col}', fontsize = 16, fontweight = 'bold', color = 'black')
    plt.xlabel(col, fontsize = 10)
    plt.ylabel('Frequency', fontsize = 10)
    plt.show()
```

Effectuons le même travail pour **сaractéristiques numériques**.

```python
# Sélectionner des caractéristiques numériques et les stocker dans une nouvelle variable
num_obj = data.select_dtypes(include=['int64', 'float64'])
print(num_obj.columns)
```

```python
# Créer des visualisations
for col in num_obj:
    plt.figure(figsize = (20, 6))
    sns.set_style('whitegrid', {'grid.color': 'lightgray', 'grid.linestyle': '--', 'grid.alpha': 0.3})

    # Créer un histogramme
    ax = sns.histplot(data[col], bins = 10, kde = True, color = 'skyblue')       # bins = 30
    # Afficher les valeurs au-dessus des colonnes
    for p in ax.patches:
        height = p.get_height()     # Hauteur de la colonne
        if height > 0:              # Vérifier que la hauteur est supérieure à zéro
            ax.text(p.get_x() + p.get_width() / 2, height, f'{int(height)}',
                    ha='center', va='bottom')
    plt.title(f'Distribution of {col}', fontsize = 16, fontweight = 'bold', color = 'black')
    plt.xlabel(col, fontsize = 10)
    plt.ylabel('Number of people', fontsize = 10)
    plt.show()
```

Construisons une matrice de corrélation pour montrer la relation entre:
* **les variables catégorielles**
* **Purchased Bike**

```python
# Codage des variables catégorielles
data_encoded = pd.get_dummies(cat_obj, drop_first = True)

# Calculer les matrices de corrélation des caractéristiques catégorielles
correlation_matrix = (data_encoded.corr()['Purchased Bike_Yes'].sort_values(ascending = False)).round(2)
correlation_matrix = correlation_matrix.reset_index()
correlation_matrix.columns = ['Column', 'Correlation']
print(correlation_matrix)

# Personnalisation de l'en-tête et des légendes
plt.figure(figsize = (20, 6))
sns.barplot(data = correlation_matrix, x = 'Column', y = 'Correlation', hue = 'Column', palette = 'RdYlBu')
plt.title('Correlation of Purchased Bikes per Column(Object)', fontsize = 16, fontweight = 'bold')
plt.xlabel('Columns(Object)', fontsize = 12)
plt.ylabel('Сorr', fontsize = 12)
plt.xticks(rotation = 90)

# Afficher les valeurs au-dessus des colonnes
for index, value in enumerate(correlation_matrix['Correlation']):
    plt.text(index, value, f'{value:.2f}', ha = 'center', va = 'bottom')

plt.tight_layout()
plt.show()
```

Construisons une matrice de corrélation pour montrer la relation entre:

* **les variables numériques**
* **Purchased Bike**


```python
# Codage des variables catégorielles
num_obj['Purchased Bike'] = pd.get_dummies(data['Purchased Bike'], drop_first = True, dtype = int)

# Calculer les matrices de corrélation des caractéristiques catégorielles
correlation_matrix = (num_obj.corr()['Purchased Bike'].sort_values(ascending = False)).round(2)
correlation_matrix = correlation_matrix.reset_index()
correlation_matrix.columns = ['Column', 'Correlation']
print(correlation_matrix)

# Personnalisation de l'en-tête et des légendes
plt.figure(figsize = (10, 6))
sns.barplot(data = correlation_matrix, x = 'Column', y = 'Correlation', hue = 'Column', palette = 'RdYlBu')
plt.title('Correlation of Purchased Bikes per Column(int)', fontsize = 16, fontweight = 'bold')
plt.xlabel('Columns(int)', fontsize = 12)
plt.ylabel('Сorr', fontsize = 12)

# Afficher les valeurs au-dessus des colonnes
for index, value in enumerate(correlation_matrix['Correlation']):
    plt.text(index, value, f'{value:.2f}', ha = 'center', va = 'bottom')

plt.tight_layout()
plt.show()
```
